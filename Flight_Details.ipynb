```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the CSV file into a Pandas DataFrame
df = pd.read_csv('/content/PPS PROJECT.csv')  # Using a raw string

def mainmenu():
    print("-----------------------")
    print("         Main Menu         ")
    print("-----------------------")
    print("1. Display The DataFrame")
    print("2. Data Analysis")
    print("3. Visualising The Data")
    print("4. Exit")
    ch = int(input("Enter your choice between 1 to 4: "))
    while ch >= 1 and ch <= 3:
        if ch == 1:
            DF()
        elif ch == 2:
            Ext()
        elif ch == 3:
            graphs()
        ch = int(input("Enter your choice between 1 to 4: "))

def DF():
    print(df)

def Ext():
    print("The minimum value of price is", df.PRICE.min())
    print("The maximum value of price is", df.PRICE.max())
    print("The maximum number of flights is", df.FLIGHTS.max())
    print("The minimum number of flights is", df.FLIGHTS.min())
    print("The maximum number of stops is", df.STOPS.max())
    print("The minimum number of stops is", df.STOPS.min())

def graphs():
    print("---------------------------")
    print("Graph Menu")
    print("---------------------------")
    print("1. Line Graph")
    print("2. Bar Vertical")
    print("3. Exit")
    chi = int(input("Enter your choice of graph between 1 and 3: "))
    while chi >= 1 and chi <= 2:  # Corrected the condition
        if chi == 1:
            line()
        elif chi == 2:
            barv()
        else:
            print("Invalid!")
        chi = int(input("Enter your choice of graph between 1 and 3: "))

def line():
    df.plot(kind="line", x="FL_CODE", y=["FLIGHTS", "STOPS"], marker='o', markersize=10,
            markerfacecolor='r', markeredgecolor='y', linewidth=4)
    plt.xticks(fontsize=12, rotation=50)
    plt.yticks(fontsize=16, rotation=8)
    plt.xlabel("FLIGHT CODE", fontsize=16)
    plt.ylabel("NO_FLIGHTS AND STOPS", fontsize=16)
    plt.legend(loc="upper left")
    plt.title("FLIGHT DETAILS", fontsize=16)
    plt.grid(True)
    plt.show()

def barv():
    df.plot(kind="bar", x="START", y="PRICE", color="orange", edgecolor="k", linewidth=5)
    plt.xticks(np.arange(0, 10), fontsize=14, rotation=90)
    plt.yticks(np.arange(5000, 16000, 1008), fontsize=12, rotation=8)
    plt.xlabel("NO STOPS", fontsize=15)
    plt.ylabel("PRICE", fontsize=16)
    plt.grid(True)
    plt.title('FLIGHT DETAILS', fontsize=16)  # Corrected title string
    plt.show()

if __name__ == "__main__":
    mainmenu()
```
